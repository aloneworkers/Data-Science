## Import dan Load Dataset

```python
from sklearn.datasets import load_digits # untuk import load_breast_cancer dengan memanggil from sklearn.datasets import load_breast_cancer

digits = load_digits() # untuk memanggil dataset dengan cara memanggil load_digits dan ditampung ke dalam variable digits
digits # untuk menampilkan isi dari variable digits
```

```python
print(digits.DESCR) # untuk mengakses deskripsi dari data sample
```

## Explanatory & Response Variables (Feature & Target)

### Explanatory Variables (Features)

```python
X = digits.data # cara mengakses features dengan memanggil breast_cancer.data
# X.shape # untuk mengakses dimensi dari datanya
X # untuk melihat isi datanya
```

Digits datasets memiliki dimensi data sejumlah 1797 baris dan 64 kolom

### Response Variables (Target)

```python
y = digits.target # untuk mengakses target dengan memanggil nama breast_cancer.target
# y.shape  # untuk menampilkan dimensinya
y #untuk melihat isi datanya
```

Digits datasets memiliki dimensi data sejumlah 1797 baris dan 1 kolom

## Visualisasi Data

```python
import matplotlib.pyplot as plt # untuk import menjadi plt

X = X[:, :2] # untuk menvisualisasikan 2 feature pertama maka dilakukan slicing

x_min, x_max = X[:, 0].min() - 0.5, X[:, 0].max() + 0.5 # x_min : nilai terkecil dari kolom indeks ke 0, x_max nilai terbesar dari kolom indeks ke 0 atau berasosiasi dengan kolom sepal length
y_min, y_max = X[:, 1].min() - 0.5, X[:, 1].max() + 0.5 # y_min : nilai terekcil dari kolom indeks ke 1, y_max nilai terbesar dari kolom indeks ke 1 atau berasosiasi dengan kolom sepal width

# untuk membentuk scatter plot
plt.scatter(X[:, 0], X[:, 1], c=y) # sumbu x diambil dari indeks ke 0 atau sepal length, sumbu y diambil dari indeks ke 1 atau sepal width, parameter c adalah color ditentukan dari nilai variable y dan nilai variable y akan berasosiasi dengan nilai targetnya
plt.xlabel('Pixel_0_0') # untuk memberikan keterangan tambahan
plt.ylabel('Pixel_4_0') # untuk memberikan keterangan tambahan

plt.xlim(x_min, x_max) # set nilai minimum dan maksimum untuk x axis
plt.ylim(y_min, y_max) # set nilai minimum dan maksimum untuk y axis
plt.grid(True) # untuk mengaktifkan grid
plt.show() # untuk memunculkan plotting
```

## Training & Testing Dataset

```python
from sklearn.model_selection import train_test_split # untuk import menjadi train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    test_size=0.3, # test set : 30%, train set : 70%
                                                    random_state=1) # untuk melakukan splitting dataset dengan cara memanggil fungsi train_test_split akan membutuhkan 4 parameter yaitu features, target atau target label, test_size akan merepresentasikan ukuran testing set dari 0 sampai 1 dan merepresentasikan persentase (%), proses splitting dilakukan dengan melakukan pengacakan agar proses pengacakan konsisten sehingga bisa melakukan replikasi perlu menentukan random_state number atau random_state
```
